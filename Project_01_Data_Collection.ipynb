## Data Collection

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
Stock_names = []
Current_price = []
Data = []
Net_Profit_margin = []
EBITDA_margin = []
Rev_24 = []
Rev_25 = []
Revenue_growth = []
Profit_growth = []



request_header = {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/140.0.0.0 Safari/537.36"
}

for page in range(0,33):
    url = f"https://groww.in/stocks/filter?closePriceHigh=1000&closePriceLow=0&marketCapHigh=3000000&marketCapLow=5000&page={page}&size=15&sortType=ASC"
    response = requests.get(url)
    #print (f"Status code {page} :",response.status_code)
    soup = BeautifulSoup(response.content,"html.parser")
    links = soup.find_all("a",class_ = "cur-po contentPrimary", href=True)

    for i in links:
        stock_link_name = i.get("href")
        
## Stock page:
        remove = ["/stocks/aditya-birla-lifestyle-brands-ltd","/stocks/jsw-cement-ltd","/stocks/tata-motors-dvr-ordinary","/stocks/schloss-bangalore-ltd",
                  "/stocks/sri-lotus-developers-and-realty-ltd","/stocks/bluestone-jewellery-and-lifestyle-ltd","/stocks/kalpataru-ltd",
                  "/stocks/max-estates-ltd","/stocks/smartworks-coworking-spaces-ltd","/stocks/saatvik-green-energy-ltd",
                  "/stocks/crizac-ltd","/stocks/vikram-solar-ltd","/stocks/jain-resource-recycling-ltd","/stocks/seshaasai-technologies-ltd",
                 "/stocks/anthem-biosciences-ltd","/stocks/urban-company-ltd","/stocks/atlanta-electricals-ltd"]
        
        if stock_link_name not in remove :
            
            stock_url = f"https://groww.in{stock_link_name}"
            response2 = requests.get(stock_url)
            soup2 = BeautifulSoup(response2.content,"html.parser")
            
            stock_name = soup2.find_all("h1",class_ = "lpu38Head truncate displaySmall")
            for name in stock_name:
                Stock_names.append(name.text.strip())
                
            cur_price = soup2.find_all("div",class_ = "lpu38Pri valign-wrapper false displayBase")
            for price in cur_price:
                Current_price.append(price.text.strip("₹"))
    
            Fundamentals = soup2.find_all("td",class_="ft785Value right-align contentPrimary bodyLargeHeavy")
            for data in Fundamentals:
                Data.append(data.text.strip())
## Income Statement:

        if stock_link_name not in remove :
            
            stock_finance_url = f"https://groww.in{stock_link_name}/company-financial/income-statement"
            response3 = requests.get(stock_finance_url)
            soup_income_statement = BeautifulSoup(response3.content,"html.parser")
            income_statement = soup_income_statement.find("tbody")
            
            npm = ebit = rev_Jun24 = rev_Jun25 = prof_24 = prof_25  = np.nan
            if income_statement:
                inc_stat = income_statement.find_all("td",class_="tb10Td")
                if len(inc_stat) >= 20 :
                    npm = inc_stat[39].text.strip()
                    print(f"{page} and {stock_link_name}")
                    ebitda = inc_stat[19].text.strip()
                    rev_Jun24 = inc_stat[0].text.strip()
                    rev_Jun25 = inc_stat[4].text.strip()
                    prof_24 = inc_stat[25].text.strip()
                    prof_25 = inc_stat[29].text.strip()
                    
                    rev_growth = ((float(rev_Jun25.replace(",","")) - float(rev_Jun24.replace(",",""))) / float(rev_Jun24.replace(",","")) ) * 100
                    prof_growth = ((float(prof_25.replace(",","")) - float(prof_24.replace(",",""))) / float(prof_24.replace(",","")) ) * 100
                    ebitda_margin = (float(ebitda.replace(",","")) / float(rev_Jun25.replace(",","")) ) * 100
            
            Net_Profit_margin.append(npm)
            EBITDA_margin.append(ebitda_margin)
            Revenue_growth.append(rev_growth)
            Profit_growth.append(prof_growth)
        

```

```python
## Extract Fundamentals Data

Market_cap = []
ROE = []
PE_Ratio = []
EPS = []
PB_ratio = []
Dividend_yield = []
Industry_PE = []
Debt_to_Equity = []

num = 0  
total_entries = 4500

for n in range(0,10):
    if n==0:
        num = n
        while (num<total_entries):
            Market_cap.append(Data[num])
            num +=10
    elif n==1:
        num = n
        while(num<total_entries):
            ROE.append(Data[num])
            num += 10
    elif n==2:
        num = n
        while(num<total_entries):
            PE_Ratio.append(Data[num])
            num += 10
    elif n==3:
        num = n
        while(num<total_entries):
            EPS.append(Data[num])
            num += 10
    elif n==4:
        num = n
        while(num<total_entries):
            PB_ratio.append(Data[num])
            num += 10
    elif n==5:
        num = n
        while(num<total_entries):
            Dividend_yield.append(Data[num])
            num += 10
    elif n==6:
        num = n
        while(num<total_entries):
            Industry_PE.append(Data[num])
            num += 10

    elif n == 8:
        num = n
        while(num<total_entries):
            Debt_to_Equity.append(Data[num])
            num += 10

            
        

```

```python
## Creating DataFrame

Stock_Df = pd.DataFrame(  {
    "Stock_Name" : Stock_names,
    "Current_Price (₹)" : Current_price,
    "Market Cap" : Market_cap,
    "Return_On_Equity (ROE) %" : ROE,
    "Price_to_Earnings_ratio (P/E)" : PE_Ratio,
    "Industry P/E" : Industry_PE,
    "Earnings_per_share (EPS)" : EPS,
    "Price_to_Book_ratio (P/B)" : PB_ratio,
    "Dividend Yield %" : Dividend_yield,
    "Debt_to_Equity" : Debt_to_Equity,
    "Revenue_Growth %" : Revenue_growth,
    "Profit_Growth %" : Profit_growth,
    "Net_Profit_Margin %" : Net_Profit_margin,
    "EBITDA Margin %" : EBITDA_margin,
}  )
```
