### Current_Price - column - Data Cleaning

```python
Stock_df["Current_Price (₹)"] = Stock_df["Current_Price (₹)"].str.replace(",","").astype("float")
```

```python
Stock_df.head()
```

Since `current price` have decimal values, changed its data type to `float`

### Market Cap - column - Data Cleaning

```python
Stock_df["Market Cap"] = Stock_df["Market Cap"].str.strip("₹Cr")
```

```python
Stock_df["Market Cap"] = Stock_df["Market Cap"].str.replace(",","").astype("int")
```

```python
Stock_df.head(2)
```

Since `Market Cap` does not have decimal values, changed its dtype to `int`

It will be easy to identify the market cap of the stocks if they are given as **Large or Mid cap**

```python
Stock_df["Market Cap"] = Stock_df["Market Cap"].apply(lambda x: "Large cap" if x > 20000 else "MId cap")
```

```python
Stock_df.tail()
```

```python
Stock_df["Market Cap"].value_counts()
```

### Return on Equity

**ROE** is percentage value and the decimal value is important for the calculation so keeping this as **float**.

```python
Stock_df["Return_On_Equity (ROE) %"] = Stock_df["Return_On_Equity (ROE) %"].str.strip("%").astype("float")
```

```python
Stock_df.head(2)
```

#### Dividend Yield

```python
Stock_df["Dividend Yield %"] = Stock_df["Dividend Yield %"].str.strip("%").astype("float")
```
